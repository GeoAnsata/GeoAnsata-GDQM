```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# 1. Ativar suporte KML no GeoPandas
gpd.io.file.fiona.drvsupport.supported_drivers['kml'] = 'rw'
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'

# 2. Caminho do arquivo KML
caminho_kml = "TESTE_GSM.kml"

# 3. Ler o arquivo KML no GeoPandas
mapa_base = gpd.read_file(caminho_kml)

# Verificar e reprojetar para UTM Zona 23S se necessrio
if mapa_base.crs != "EPSG:32723":  
    mapa_base = mapa_base.to_crs(epsg=32723)

# 4. Carregar o DataFrame com as coordenadas
# Substitua o caminho pelo seu arquivo CSV ou outro formato
caminho_csv = "Furos de Sonda AVANTE_Headers.csv"  # Seu arquivo de pontos
df = pd.read_csv(caminho_csv)

# Verificar se as colunas 'coordinates.X' e 'coordinates.Y' existem
if "coordinates.X" in df.columns and "coordinates.Y" in df.columns:
    print("Colunas de coordenadas encontradas.")
else:
    raise ValueError("Colunas 'coordinates.X' e 'coordinates.Y' no encontradas no DataFrame.")

# 5. Criar GeoDataFrame a partir das coordenadas
gdf_pontos = gpd.GeoDataFrame(
    df, 
    geometry=[Point(x, y) for x, y in zip(df["coordinates.X"], df["coordinates.Y"])]
)
gdf_pontos.set_crs(epsg=32723, inplace=True)  # Definir sistema de coordenadas UTM Zona 23S

# 6. Plotar o mapa base com os pontos
fig, ax = plt.subplots(figsize=(12, 8))

# Plotar o mapa base
mapa_base.plot(ax=ax, color="lightgray", edgecolor="black", alpha=0.5, label="Mapa Base")

# Plotar os pontos
gdf_pontos.plot(ax=ax, color="red", markersize=50, label="Pontos do DataFrame")

# Personalizar o grfico
plt.title("Pontos UTM Sobre Mapa Base KML (Zona 23S)")
plt.xlabel("Coordenada X (UTM)")
plt.ylabel("Coordenada Y (UTM)")
plt.legend()
plt.grid(visible=True, linestyle="--", linewidth=0.5)
plt.show()
```
